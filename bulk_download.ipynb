```python
import ftplib
import os

def download_files(ftp, path, local_path):

    try:
        ftp.cwd(path)
        os.makedirs(local_path, exist_ok=True)
        print(f"Changed to {path}")

        items = ftp.nlst()

        for item in items:
            try:
                ftp.cwd(f"{path}/{item}")
                is_directory = True
                ftp.cwd(path)

            except ftplib.error_perm:
                is_directory = False

            if is_directory:
                download_files(ftp, f"{path}/{item}", os.path.join(local_path, item))

            else:
                with open(os.path.join(local_path, item), 'wb') as local_file:
                    ftp.retrbinary('RETR ' + item, local_file.write)
                    print(f"Downloaded {item} to {local_path}")

    except Exception as e:
        print(f"Error: {str(e)}")


ftp_server = "ftp.ceda.ac.uk"

# desired folder to download
remote_path = "/badc/deposited2021/adverse_met_scenarios_electricity/data/long_duration/"

# location for the download
local_path = "C:/Users/pz61/Box/CIReN_Shared/Task 2a - climate impacts/Data/"

# FTP login details
ftp_user = "pz61"
ftp_password = ".Du>l1+7Nm'%"


ftp = ftplib.FTP(ftp_server)

ftp.login(ftp_user, ftp_password)

download_files(ftp, remote_path, local_path)

ftp.quit()
```
